```python
# terminal: run 'ipcluster start --n=<no_engines>
from venture.venturemagics.ip_parallel import *
import numpy as np
import matplotlib.pylab as plt

lite = False # choose backend
clear_all_engines() # clear ipcluster engines
```

```python
regression='''
[assume w0 (normal 0 3) ]
[assume w1 (normal 0 3) ]
[assume omega (normal 0 3) ]
[assume theta (normal 0 3) ]
[assume x (mem (lambda (i) (normal 0 5) ) )]
[assume noise (gamma 1 1) ]

[assume model (if (flip) 1 0) ]
[assume linear (lambda (x) (+ w0 (* w1 x) ) ) ]
[assume fourier (lambda (x) (+ w0 (* w1 (sin (+ (* omega x) theta) ) ) ) ) ]
[assume f (if (= model 0) linear fourier) ]

[assume y (mem (lambda (i) (normal (f (x i) ) noise) ) )]
'''
```

```python
# start with a small number of ripls to explore the prior
v=MRipl(5,lite=lite,verbose=False)
v.execute_program(regression);
```

```python
%%mr_map v plot_f store_prior
import numpy as np
no_sample_points=20

def plot_f(ripl):
    
    # pull params from ripl
    params = ['model','w0','w1','omega','theta','noise']
    inf_params = {varname:ripl.sample(varname) for varname in params}
    model,w0,w1,omega,theta,noise = [inf_params[key] for key in params]
    
    if model==0: 
        f=lambda x,w0,w1,omega,theta: w0+(w1*x)
    else:
        f=lambda x,w0,w1,omega,theta: w0 + ( w1*(np.sin(omega*x + theta) ) )
    #FIXME PICKLE ERROR inf_params['f']=f
    
    # pull some x,y pairs from ripl
    xs = [ripl.predict('(x %i)' % i) for i in range(no_sample_points) ]
    ys = [ripl.predict('(y %i)' % i) for i in range(no_sample_points) ]
    xr = np.linspace(min(xs),max(xs),100)
    
    # plot function f and some x,y pairs
    fig,ax = plt.subplots(1,2,figsize=(8,2))
    ax[0].scatter(xs,ys)
    ax[1].plot(xr,[f(x,w0,w1,omega,theta) for x in xr])
    
    return inf_params
```

```python
# plot some of the functions on same axis
inf_params_list = store_prior['out']

max_funcs_to_plot = 5; limit = min(len(inf_params_list),max_funcs_to_plot)

fig,ax = subplots()
xr = np.linspace(-10,10,100)

for inf_params in inf_params_list[:limit]:
    model,w0,w1,omega,theta,noise = [inf_params[key] for key in ['model','w0','w1','omega','theta','noise']]

    if model==0: 
        f=lambda x,w0,w1,omega,theta: w0+(w1*x)
    else:
        f=lambda x,w0,w1,omega,theta: w0 + ( w1*(np.sin(omega*x + theta) ) )

    ax.plot(xr,[f(x,w0,w1,omega,theta) for x in xr])
    
ym = ax.get_ylim();
ax.set_ylim([.6*ym[0],.6*ym[1]]);
```

```python
# now work with a larger number of ripls
v=MRipl(20,lite=lite,verbose=False)
v.execute_program(regression);

# generate synthetic data
n=20
xs = np.concatenate( (np.random.normal(loc=-1,scale=1,size=n/2), np.random.normal(loc=0.5,scale=2,size=n/2)) )
ys = 2*np.sin(xs)-.5
xys = zip(xs,ys)

fig,ax = subplots(figsize=(7,3))
ax.scatter(xs,ys)

for i,(x,y) in enumerate(xys):
    v.observe('(x %i)' % i , '%f' % x) 
    v.observe('(y %i)' % i , '%f' % y)
```

```python
# look at snapshots before and after
prior_w = v.snapshot(exp_list=['w0','w1'],plot=False,scatter=True);
prior_noise = v.snapshot(exp_list=['noise'],plot=True)
```

```python
v.infer(100)
post_w = v.snapshot(exp_list=['w0','w1'],plot=False,scatter=True); # generative model: w0=-.5, w1=2
post_noise = v.snapshot(exp_list=['noise'],plot=True)
```
