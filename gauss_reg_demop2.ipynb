```python
%%px
def get_name(r_mr):
    try:
        r_mr.dview; mr=1
    except:
        mr=0
    di_l = r_mr.list_directives()[0] if mr else r_mr.list_directives()
    if 'model_name' in str(di_l):
        try:
            n = r_mr.sample('model_name')[0] if mr else r_mr.sample('model_name')
            return n
        except: pass
    else:
        return 'anon model'
    
import numpy as np    
from scipy.stats import kde

def heatplot(n2array,nbins=100):
    """Input is an nx2 array, plots graph and returns xi,yi,zi for colormesh""" 
    x, y = n2array.T
    k = kde.gaussian_kde(n2array.T)
    xi, yi = np.mgrid[x.min():x.max():nbins*1j, y.min():y.max():nbins*1j]
    zi = k(np.vstack([xi.flatten(), yi.flatten()]))
    # plot ax.pcolormesh(xi, yi, zi.reshape(xi.shape))
    return (xi, yi, zi.reshape(xi.shape))

def if_lst_flatten(l):
    if type(l[0])==list: return [el for subl in l for el in subl]
    return l
```

```python
male18=[179,195.1,183.7,178.7,171.5,181.8,
        172.5,174.6,190.4,173.8,172.6,
        185.2,178.4,177.6,183.5,178.1,177,
        172.9,188.4,169.4,180.2,189,182.4,
        185.8,180.7,178.7]
data=male18
```

```python
def plot_gauss_infer():
    v.snapshot(exp_list=['mu','sigma'],plot=True,scatter=True)
    snaps=[v.snapshot(exp_list=['(normal mu sigma)'],plot=False)['values']['(normal mu sigma)'] for i in range(20)]
    xs = if_lst_flatten(snaps)
    fig,ax = subplots(1,2,figsize=(12,3))
    ax[0].hist(xs,bins=20,color='m')
    ax[0].set_title('Posterior predictive distribution on height')
    xr = np.linspace(min(xs),max(xs),100)
    ax[1].plot(xr,kde.gaussian_kde(xs)(xr),c='m')
```

## Function learning

```python
v=MRipl(4)
```

suppose we want to learn a function from a series of (x,y) pairs
the function f is known to be constant but there's measurement noise

```python
%%venture v
[clear]
[assume x (lambda () (uniform_continuous -3 3))]

[assume w0 (normal 0 10) ]
[assume f (lambda (x) w0 ) ]
[assume noise (gamma 5 1) ]

[assume y (lambda (x) (normal (f x) noise) ) ]

[assume model_name (quote constant)]
```

inferring the constant function f is essentially the same
as inferring a gaussian's parameters. we observe (x,y) points instead of 
1D scalars x, but the x in the (x,y) is irrelevant

```python
xs = normal(0,1,6); ys=normal(8,1,6)
xys=zip(xs,ys)
fig,ax=subplots(figsize=(4,2))
ax.scatter(xs,ys)
ax.set_title('Data from constant f with Gaussian noise');
ax.set_ylim(0,25)
print 'xs= ',xs; print 'ys= ',ys
```

```python
[v.observe('(y %f)' %x,str(y) ) for x,y in xys]
v.infer(200)
```

```python
#%px xs=  [-0.60103092  0.85039239  0.01959949  1.12179003  0.04520985  0.721185  ]
#%px ys=  [ 6.33300895  6.03576078  8.57150023  6.88232394  8.77649265  7.39969793]
v.dview.push({'xs':xs,'ys':ys});
```

```python
def plot_prior(ripl):
    xr=np.linspace(-3,3,40)
    f_xr=[ripl.sample('(f %f)' % x) for x in xr]
    f_a = [fx+ripl.sample('noise') for fx in f_xr]
    f_b = [fx-ripl.sample('noise') for fx in f_xr]
    
    fig,ax = plt.subplots(figsize=(4,2))   
    ax.set_color_cycle(['m', 'gray','gray'])
    ax.plot(xr,f_xr,xr,f_a,xr,f_b)
    ax.set_title('Ripl: f (+- 1sd)')

def plot_f(ripl):
    xr=np.linspace(1.5*min(xs),1.5*max(xs),40)
    f_xr=[ripl.sample('(f %f)' % x) for x in xr]
    f_a = [fx+ripl.sample('noise') for fx in f_xr]
    f_b = [fx-ripl.sample('noise') for fx in f_xr]
    
    fig,ax = plt.subplots(figsize=(4,2))   
    ax.set_color_cycle(['m', 'gray','gray'])
    ax.plot(xr,f_xr,xr,f_a,xr,f_b)
    ax.set_title('Ripl: f (+- 1sd)')
    ax.scatter(xs,ys)
    
out = mr_map_f(v,plot_f,limit=1)
```

```python
def plot_conditional(ripl):
    xr=np.linspace(1.5*min(xs),1.5*max(xs),40)
    f_xr=[ripl.sample('(f %f)' % x) for x in xr]
    f_a = [fx+ripl.sample('noise') for fx in f_xr]
    f_b = [fx-ripl.sample('noise') for fx in f_xr]
    
    fig,ax = plt.subplots(1,3,figsize=(13,3),sharex=True,sharey=True)
    
    ax[0].set_color_cycle(['m', 'gray','gray'])
    ax[0].plot(xr,f_xr,xr,f_a,xr,f_b)
    ax[0].set_title('Ripl: f (+- 1sd)')
    ax[0].scatter(xs,ys)
    
    xys = np.array([(x,ripl.sample('(y %f)'%x)) for x in xr for j in range(50)])
    ax[1].scatter(xys[:,0],xys[:,1],s=5,c='gray')  
    ax[1].set_title('Ripl: Scatter P(y/X=x,params)')
        
    xi,yi,zi=heatplot(xys,nbins=100)
    ax[2].pcolormesh(xi, yi, zi)
    ax[2].set_title('Ripl: GKDE P(y/X=x,params)')
    fig.tight_layout()
```
