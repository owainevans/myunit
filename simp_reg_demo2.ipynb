```python
## we now use two mripls to compare two different models of the same data set

from venture.venturemagics.ip_parallel import *; lite=False; clear_all_engines()

quadratic='''
[assume w0 (normal 0 3) ]
[assume w1 (normal 0 3) ]
[assume w2 (normal 0 3) ]
[assume x (mem (lambda (i) (normal 0. 5.) ) )]
[assume noise (gamma 1 1) ]
[assume f (lambda (x) (+ w0 (* w1 x) (* w2 (* x x)) ) ) ]
[assume y (mem (lambda (i) (normal (f (x i) ) noise) ) )]
'''
fourier='''
[assume w0 (normal 0 3) ]
[assume w1 (normal 0 3) ]
[assume omega (normal 0 3) ]
[assume theta (normal 0 3) ]
[assume x (mem (lambda (i) (normal 0. 5.) ) )]
[assume noise (gamma 1 1) ]
[assume f (lambda (x) (+ w0 (* w1 (sin (+ (* omega x) theta) ) ) ) ) ]
[assume y (mem (lambda (i) (normal (f (x i) ) noise) ) )]
'''

vq = MRipl(2,lite=lite); vq.execute_program(quadratic)
vf = MRipl(2,lite=lite); vf.execute_program(fourier)
vs=[vq,vf]


def generate_sin_data(n):
    xs = np.random.normal(loc=0,scale=4,size=n)
    ys = 3*np.sin(xs)
    xys = zip(xs,ys)
    fig,ax = subplots(figsize=(6,2))
    ax.scatter(xs,ys); ax.set_title('Data for regression problem')
    return xys

xys = generate_sin_data(6)

# load data as observation for each mripl and do some inference
no_transitions=200
for i,(x,y) in enumerate(xys):
    [v.observe('(x %i)' % i , '%f' % x) for v in vs]
    [v.observe('(y %i)' % i , '%f' % y) for v in vs]
    
[v.infer(no_transitions) for v in vs]
```

```python
%%mr_map vq plot_curve store_plot_q
import numpy as np
n=6 

def plot_curve(ripl):
    xs = [ripl.sample('(x %i)' % i) for i in range(n)]
    ys = [ripl.sample('(y %i)' % i) for i in range(n)]
    xr = np.linspace(2*min(xs),2*max(xs),100)
    f_xr = [ripl.sample('(f %f)' % x) for x in xr]
    
    noise = ripl.sample('noise')
    f_a = [fx+noise for fx in f_xr]
    f_b = [fx-noise for fx in f_xr]
    
    fig,ax = plt.subplots(figsize=(8,3))
    ax.scatter(xs,ys)
    ax.plot(xr,f_xr,c='red')
    ax.plot(xr,f_a,c=.6); ax.plot(xr,f_b,c=.6)
    return None
```

```python
## Posterior samples from the models

no_ripls = 5
vq = MRipl(no_ripls,lite=lite,verbose=False); vq.execute_program(quadratic)
vf = MRipl(no_ripls,lite=lite,verbose=False); vf.execute_program(fourier)
vs=[vq,vf]
no_transitions=500
xys = generate_sin_data(15); xs = [pair[0] for pair in xys]

for i,(x,y) in enumerate(xys):
    [v.observe('(x %i)' % i , '%f' % x) for v in vs]
    [v.observe('(y %i)' % i , '%f' % y) for v in vs]
xr = np.linspace(1.3*min(xs),1.3*max(xs),30)

[v.infer(no_transitions) for v in vs]

y_vals = [ [v.sample('(normal (f %f) noise)' % x) for x in xr] for v in vs]
```
