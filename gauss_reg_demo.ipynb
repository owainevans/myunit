# Inferring the Mean and Variance of a Gaussian Distribution

These are the heights of a group of a 18-year-old men. We'd like to make an inference about the population based on this data.

```python
male18=[179,195.1,183.7,178.7,171.5,181.8,172.5,
        174.6,190.4,173.8,172.6,185.2,178.4,
        177.6,183.5,178.1,177,172.9,188.4,169.4,
        180.2,189,182.4,185.8,180.7,178.7]
data=male18
fig,ax=subplots(figsize=(4,2))
ax.hist(data,color='orange'); ax.set_title('Heights of 18 y/o Males');
```

We model the population distribution as a Gaussian with unknown mean and variance.

```python
v=MRipl(10)
#try bigger numbers
```

```python
%%venture v
[assume mu (normal 0 60) ]
[assume sigma (gamma 2 .2) ]
[assume x (lambda () (normal mu sigma) ) ]
```

```python
v.snapshot(exp_list=['mu','sigma'],plot=True,scatter=True);
```

Each pair of mu and sigma produces a distinct Gaussian population distribution. Here are some sample synthetic population distributions from our prior.

```python
out=v.snapshot(exp_list='(x)',plot=True,sample_populations=(5,20));
# sample_populations=(no_populations,population_size)
# should be slow for large pop size
```

```python
out=v.snapshot(exp_list='(x)',plot=True,sample_populations=(4,90));
```

```python
outs=v.snapshot(exp_list='(x)',plot=True,repeat=100)
# if repeat is set, we get repeat draws from each ripl (as in '(repeat n thunk)'. i could change name
```

### Inference from data

```python
prior_mu = v.snapshot(exp_list='mu')
v.observe('(x)',str(data[0]) )
print 'Observed data: ', data[:1]
v.infer(500)
posterior1_mu=v.snapshot(exp_list='mu')
fig = v.snapshot(plot_past_values=[prior_mu,posterior1_mu])
```

```python
v.observe('(x)',str(data[1]) )
print 'Observed data: ', data[:2]
v.infer(500)
posterior2_mu=v.snapshot(exp_list='mu')
fig = v.snapshot(plot_past_values=[posterior1_mu,posterior2_mu])
```

```python
v.observe('(x)',str(data[2]) )
v.observe('(x)',str(data[3]) )
print 'Observed data: ', data[:4]
v.infer(500)
posterior3_mu=v.snapshot(exp_list='mu')
fig = v.snapshot(plot_past_values=[prior_mu,posterior1_mu,posterior2_mu,posterior3_mu])
```

## Making inference robust to outliers

```python
v=MRipl(40)
```

```python
%%venture v
[assume mu (normal 0 60) ]
[assume sigma (gamma 2 .2) ]
[assume x (lambda () (normal mu sigma))]
```

```python
[v.observe('(x)',str(data[i])) for i in range(4) ]
v.infer(600);
posterior_before_outlier=v.snapshot(exp_list='mu',plot=True);
```

```python
v.observe('(x)',str(325))
v.infer(600)
posterior_after_outlier=v.snapshot(exp_list='mu',plot=True,plot_past_values=[posterior_before_outlier])
```

```python
vr = MRipl(20)
```

```python
%%venture vr
[assume mu (normal 0 60) ]
[assume sigma (gamma 2 .2) ]
[assume x (lambda () (if (flip .96) (normal mu sigma) (normal mu (* 10 sigma) ) ) ) ]
```

```python
[v.observe('(x)',str(data[i])) for i in range(4) ]
v.infer(600);
posterior_before_outlier=v.snapshot(exp_list='mu',plot=False);
v.observe('(x)',str(325))
v.infer(600)
posterior_after_outlier=v.snapshot(exp_list='mu',plot=True,plot_past_values=[posterior_before_outlier])
```

```python
print 'basic model mean logscore: ', np.mean(v.get_global_logscore())
print 'robust model mean logscore: ', np.mean(vr.get_global_logscore())
```
