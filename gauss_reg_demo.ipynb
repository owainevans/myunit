# Mean, Variance of Gaussian

You observe a small sample of heights from some population. What can you infer about the population distribution?

```python
# heights
xs = [2.1,  1.99,  1.88,  1.94,  2.01]
plt.hist(xs);
```

We model the population distribution as a Gaussian with unknown mean and variance.

```python
v=MRipl(50,lite=lite,verbose=False)
v.assume('x','(mem (lambda (i) (normal mu sigma) ) )');
```

The procedure <code>x</code> will generate heights by calling $\mu$ and $\sigma$. These values are unknown, but we can specify prior distributions on them (based on background knowledge). 

```python
v.assume('mu','(gamma 3 2)')
v.assume('sigma','(gamma .4 1)');
```

We can visualize these prior distributions on mu and sigma. 

```python
v.snapshot(exp_list=['mu','sigma'],plot=True);
```

Every pair of possible mu,sigma values will produce different population distributions. We can also plot a set of random population distributions based on our prior.

```python
%%mr_map v plot_population_dist st 2
def plot_population_dist(ripl):
    xs=[ripl.sample('(x %i)'%index) for index in range(200)]
    mu=ripl.sample('mu'); sigma = ripl.sample('sigma')
    fig,ax = plt.subplots(figsize=(5,2))
    ax.hist(xs,bins=20)
    ax.set_title('Population distribution on heights (mu=%f, sigma=%f)' %(mu,sigma))
    return None #'Sample heights: ', xs[:5]
```

We can also plot our uncertainty over a single observed height (i.e. our distribution over a height from an unknown population).

```python
v.snapshot(exp_list=['(normal mu sigma)'],plot=True);
```

```python
# look again at heights
xs = [2.1,  1.99,  1.88,  1.94,  2.01]
# compared to prior, these are unusually large heights. we now condition on these observations
```

```python
[v.observe('(x %i)'%index, str(x)) for index,x in enumerate(xs) ];
```

```python
# we can look at our distribution over a new height drawn from the population before and after inference
out_prior=v.snapshot(exp_list=['(normal mu sigma)'],plot=True);
v.infer(100)
out_posterior=v.snapshot(exp_list=['(normal mu sigma)'],plot=True);
```

```python
# we can look again at our distributions on the mean and variance of the gaussian.
# we expect the mean to be close the sample mean, but smaller than it
out=v.snapshot(exp_list=['mu'],plot=True);
print 'posterior_mean: ',np.mean(out['values']['mu'])
print 'sample_mean: ',np.mean(xs)
```

```python
# we now add another observation
xs.append(2.4)
v.observe('(x %i)' % len(xs), str(xs[-1]))
plt.hist(xs);
```

```python
# we can look at the effect of this single point on our distribution on mu
out=v.snapshot(exp_list=['mu'],plot=True);
print 'posterior_mean: ',np.mean(out['values']['mu'])
print 'sample_mean: ',np.mean(xs)
```

```python
# suppose we want to learn a function from a series of (x,y) pairs
# the function f is known to be constant but there's measurement noise

constant_model='''
[assume x (mem (lambda (i) (uniform_continuous -3 3)))]

[assume w0 (normal 0 10) ]
[assume f (lambda (x) w0 ) ]
[assume noise (gamma 5 1) ]

[assume y_x (lambda (x) (normal (f x) noise) ) ]
[assume y (mem (lambda (i) (y_x (x i)) ) )]

[assume n (gamma 1 100)]
[assume model_name (quote constant)]'''
# suppose we want to learn a function that we know to be constant, but where there is measurement noise
v=MRipl(3,lite=lite,verbose=False); v.execute_program(constant_model);
```

```python
# we see that this problem of inferring the constant function f is essentially the same
# as the problem of inferring a gaussian. this time we observe (x,y) points instead of 
# 1D scalars x, but the x in the (x,y) is irrelevant

xys = np.array(zip(np.random.normal(0,1,6),np.random.normal(20,1,6)))
fig,ax=subplots(figsize=(4,4))
ax.scatter(xys[:,0],xys[:,1])
ax.set_title('Data from constant f with Gaussian noise');
ax.set_ylim(0,25)
# inference
no_transitions = 100
observe_infer([v],xys,no_transitions)
```

```python
# inferred functions
mr_map_nomagic(v,plot_cond,limit=1);
```

```python
# to move from a constant function to a linear function, we just add one expression to the code
linear_model='''
[assume x_d (lambda ()(uniform_continuous -3 3)) ]
[assume x (mem (lambda (i) (x_d) ))]

[assume w0 (normal 0 5) ]
[assume w1 (normal 0 5) ]
[assume f (lambda (x) (+ w0 (* w1 x) ) ) ]
[assume noise (gamma 5 1) ]

[assume y_x (lambda (x) (normal (f x) noise) ) ]
[assume y (mem (lambda (i) (y_x (x i)) ) )]

[assume n (gamma 1 100)]
[assume model_name (quote linear)]'''
```

```python
v=MRipl(3,lite=lite,verbose=False); v.execute_program(linear_model);
# we consider getting data point by point, and showing how our inference develops
xys = [(.1, .8), (.9,.7), (-1.1,-.9), (1.6,.4), (-1.9, -2.4)] 
```

```python
# first we look at some functions from the prior
mr_map_nomagic(v,plot_cond,limit=1);
```
