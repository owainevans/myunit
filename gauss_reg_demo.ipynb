# Mean, Variance of Gaussian

You observe a small sample of heights from some population. What can you infer about the population distribution?

```python
boys18=[179,195.1,183.7,178.7,171.5,181.8,172.5,174.6,190.4,173.8,172.6,185.2,178.4,177.6,183.5,178.1,177,172.9,188.4,169.4,180.2,189,182.4,185.8,180.7,178.7
]
data=boys18
plt.hist(data);
```

We model the population distribution as a Gaussian with unknown mean and variance.

```python
v=MRipl(50)
```

```python
%%venture v
[assume mu (normal 0 40) ]
[assume sigma (gamma 2 .2) ]
```

We can visualize these prior distributions on mu and sigma. 

```python
v.snapshot(exp_list=['mu','sigma'],plot=True);
```

Every pair of possible mu,sigma values will produce different population distributions. We can also plot a set of random population distributions based on our prior.

```python
def generate_synthetic_population(ripl):
    return [ripl.sample("(normal mu sigma)") for i in range(200)]

def plot_synthetic_populations():
    populations = mr_map_f(v,generate_synthetic_population,limit=2)['out']
    fig,ax = subplots(1,2,figsize=(14,5))
    [ax[0].hist(population,bins=20) for population in populations];
    
    xmin = min([min(pop) for pop in populations]); xmax = max([max(pop) for pop in populations])
    xr = np.linspace(xmin,xmax,100)
    
    for pop in populations:
        pdf = kde.gaussian_kde(pop)
        ax[1].plot(xr,pdf(xr))
plot_synthetic_populations();
```

Summarize the prior. We show distributions on mu and sigma, their relationship, and the predictive distribution on x.

```python
def plot_gauss_infer():
    v.snapshot(exp_list=['mu','sigma'],plot=True,scatter=True)
    snaps=[v.snapshot(exp_list=['(normal mu sigma)'],plot=False)['values']['(normal mu sigma)'] for i in range(20)]
    xs = if_lst_flatten(snaps)
    fig,ax = subplots(1,2,figsize=(12,3))
    ax[0].hist(xs,bins=20,color='m')
    ax[0].set_title('Posterior predictive distribution on height')
    xr = np.linspace(min(xs),max(xs),100)
    ax[1].plot(xr,kde.gaussian_kde(xs)(xr),c='m')
plot_gauss_infer()
```

Show how our distributions on mu,sigma and the distribution on heights changes as we observe data.
