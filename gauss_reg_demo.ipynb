# Inferring the Mean and Variance of a Gaussian Distribution

These are the heights of a group of a 18-year-old men. We'd like to make an inference about the population based on this data.

```python
male18=[179,195.1,183.7,178.7,171.5,181.8,172.5,
        174.6,190.4,173.8,172.6,185.2,178.4,
        177.6,183.5,178.1,177,172.9,188.4,169.4,
        180.2,189,182.4,185.8,180.7,178.7]
data=male18
fig,ax=subplots(figsize=(4,2))
ax.hist(data,color='orange'); ax.set_title('Heights of 18yo Males');
```

We model the population distribution as a Gaussian with unknown mean and variance.

```python
v=MRipl(10)
```

```python
%%venture v
[assume mu (normal 0 60) ]
[assume sigma (gamma 2 .2) ]
```

```python
v.snapshot(exp_list=['mu','sigma'],plot=True,scatter=True);
```

Each pair of mu and sigma produces a distinct Gaussian population distribution. Here are some sample synthetic population distributions from our prior.

```python
xs=[]
for i in range(4):
    xs.append( v.snapshot(exp_list='(normal mu sigma)') )
v.snapshot(exp_list='(normal mu sigma)',plot_range=(-140,140),plot_past_values=xs);
```

```python
def generate_synthetic_population(ripl):
    return [ripl.sample("(normal mu sigma)") for i in range(200)]

def plot_synthetic_populations():
    populations = multiripl_map_proc(v,generate_synthetic_population,limit=2)['out']
    fig,ax = subplots(1,2,figsize=(12,4))
    [ax[0].hist(population,bins=20) for population in populations];
    xr = np.linspace(rmin(populations),rmax(populations),100)
    
    for population in populations:
        pdf = kde.gaussian_kde(population)
        ax[1].plot(xr,pdf(xr))
        
plot_synthetic_populations();

### FIXME: can flatten list to get posterior predictive.
```

Summarize the prior. We show distributions on mu and sigma, their relationship, and the predictive distribution on x.

```python
def plot_gauss_infer():
    v.snapshot(exp_list=['mu','sigma'],plot=True,scatter=True)
    snaps=[v.snapshot(exp_list=['(normal mu sigma)'],plot=False)['values']['(normal mu sigma)'] for i in range(20)]
    xs = if_lst_flatten(snaps)
    fig,ax = subplots(1,2,figsize=(9,2))
    ax[0].hist(xs,bins=20,color='m')
    ax[0].set_title('Posterior predictive distribution on height')
    xr = np.linspace(min(xs),max(xs),100)
    ax[1].plot(xr,kde.gaussian_kde(xs)(xr),c='m')
plot_gauss_infer()
## fixme put scatter in earlier plot
```

### Inference from data

```python
prior=v.sample('mu')
v.observe('(normal mu sigma)',str(data[0]) )
print 'Observed data: ', data[:1]
v.infer(500)
posterior1=v.sample('mu')

def plot_update(prior,posterior):
    fig,ax = subplots(figsize=(3,2))
    xr = np.linspace(min(prior+posterior),max(prior+posterior),40)
    ax.plot(xr,kde.gaussian_kde(prior)(xr),c='blue',label='Prior on mu')
    ax.plot(xr,kde.gaussian_kde(posterior)(xr),c='green',label='Posterior on mu');
plot_update(prior,posterior1)
```

```python
v.observe('(normal mu sigma)',str(data[1]) )
print 'Observed data: ', data[:2]
v.infer(500)
posterior2=v.sample('mu')

plot_update(posterior1,posterior2)
```

```python
v.observe('(normal mu sigma)',str(data[2]) )
print 'Observed data: ', data[:3]
v.infer(500)
posterior3=v.sample('mu')

plot_update(posterior2,posterior3)
```
