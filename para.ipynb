```python
def wm2(no_ripls):
    v = MultiRipl(no_ripls)
    v.assume('r','(normal 0 30)')
    v.assume('s','(normal 0 30)')
    v.assume('w','(normal (+ r s) 5.)')
    v.observe('w','50.')
    v.assume('w2','(normal (+ r s) 6.)')
    v.observe('w2','50.')
    return v
```
