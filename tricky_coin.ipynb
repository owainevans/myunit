### Switching coins over time

```python
v=MRipl(10)
```

```python
%%venture v
[clear]
```

```python
%%venture v
[assume tricky_coin_weight (uniform_continuous 0 1)]
[assume pick_tricky (lambda (t) (if (= t 0) (flip)
                                 (if (pick_tricky (- t 1)) 
                                  (flip .85) (flip .3) ) ) )]
[assume flip_coin (lambda (t) (if (pick_tricky t) (flip tricky_coin_weight) (flip 0.5) ) ) ]
```

```python
%%venture v
[sample (quote WEIGHT)]
[sample tricky_coin_weight]
[sample (quote FLIPS)]
[sample (flip_coin 0) ]
[sample (flip_coin 1) ]
[sample (flip_coin 2) ]
[sample (flip_coin 3) ]
[sample (flip_coin 4) ]
[sample (flip_coin 5) ]
[sample (flip_coin 6) ]
[sample (flip_coin 7) ]
[sample (flip_coin 8) ]
[sample (flip_coin 9) ]
[sample (flip_coin 10) ]
[sample (flip_coin 11) ]
[sample (flip_coin 12) ]
[sample (flip_coin 13) ]
[sample (flip_coin 14) ]
[sample (flip_coin 15) ]
```

```python
%%venture v
[observe (flip_coin 0) true ]
[observe (flip_coin 1) true ]
[observe (flip_coin 2) true ]
[observe (flip_coin 3) true ]
[observe (flip_coin 4) true ]
```

```python
v.infer(300)
print 'After observing five Heads, P(next flip is Heads)= ', np.mean(v.sample('(flip_coin 5)'))
print 'After observing five Heads, P( tricky was picked first )= ', np.mean(v.sample('(pick_tricky 0)'))
posterior1_samples=v.snapshot(exp_list='tricky_coin_weight',plot=True)
```
